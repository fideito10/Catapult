```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
import time

def obtener_posicion_equipo_urba(nombre_equipo: str = "Universitario de La Plata") -> dict:
    """
    Devuelve la posición actual del equipo en la tabla de URBA.
    
    Args:
        nombre_equipo (str): Nombre del equipo (por defecto: Universitario de La Plata)
        
    Returns:
        dict: {'equipo': nombre, 'posicion': int, 'text': texto_completo} o None si no se encuentra
    """
    
    url = "https://fixture.urba.org.ar/table?id=2025008"
    
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Para que no abra ventana (modo silencioso)
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    time.sleep(6)

    equipos = driver.find_elements(By.XPATH, "//div[contains(@class, 'teamWrapper')]")

    for equipo in equipos:
        texto = equipo.text.lower()
        if nombre_equipo.lower() in texto:
            # Extraer posición (está en el <h5> dentro del div)
            posicion_elem = equipo.find_element(By.TAG_NAME, "h5")
            posicion = int(posicion_elem.text.strip())
            
            driver.quit()
            return {
                "equipo": nombre_equipo,
                "posicion": posicion,
                "texto": equipo.text
            }

    driver.quit()
    return None  # Si no se encuentra el equipo
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

def obtener_posicion_equipo_urba(nombre_equipo: str = "Universitario de La Plata") -> dict:
    """
    Devuelve la posición actual del equipo en la tabla de URBA.
    
    Args:
        nombre_equipo (str): Nombre del equipo (por defecto: Universitario de La Plata)
        
    Returns:
        dict: {'equipo': nombre, 'posicion': int, 'text': texto_completo} o None si no se encuentra
    """
    
    url = "https://fixture.urba.org.ar/table?id=2025008"
    
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--window-size=1920,1080')  # Asegurar ventana grande
    
    driver = None
    
    try:
        driver = webdriver.Chrome(options=options)
        driver.get(url)
        
        print("Esperando a que la página cargue...")
        
        # Esperar explícitamente hasta que aparezca algún elemento con clase teamWrapper
        wait = WebDriverWait(driver, 15)  # Aumentamos el tiempo de espera a 15 segundos
        wait.until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'teamWrapper')]")))
        
        print("Página cargada, buscando equipos...")
        
        # Hacer scroll para asegurar que todos los elementos estén cargados
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Breve pausa después del scroll
        
        equipos = driver.find_elements(By.XPATH, "//div[contains(@class, 'teamWrapper')]")
        print(f"Se encontraron {len(equipos)} equipos en la tabla")
        
        # Intentar con diferentes selectores si no se encontraron equipos
        if len(equipos) == 0:
            print("Probando con selector alternativo...")
            equipos = driver.find_elements(By.XPATH, "//div[contains(@class, 'team')]")
            print(f"Se encontraron {len(equipos)} equipos con selector alternativo")
        
        for i, equipo in enumerate(equipos):
            texto = equipo.text.lower()
            print(f"Equipo {i+1}: {texto}")
            
            # Intentar con diferentes variantes del nombre
            nombre_equipo_lower = nombre_equipo.lower()
            if (nombre_equipo_lower in texto or 
                "universitario" in texto and "plata" in texto):
                print(f"¡Equipo encontrado! Texto: {equipo.text}")
                
                try:
                    # Extraer posición (primero intentamos con h5)
                    posicion_elem = None
                    try:
                        posicion_elem = equipo.find_element(By.TAG_NAME, "h5")
                    except:
                        # Si no encuentra h5, busca cualquier elemento con un número
                        elementos = equipo.find_elements(By.XPATH, ".//*")
                        for elem in elementos:
                            if elem.text.strip().isdigit():
                                posicion_elem = elem
                                break
                    
                    if posicion_elem:
                        posicion = int(posicion_elem.text.strip())
                        
                        return {
                            "equipo": nombre_equipo,
                            "posicion": posicion,
                            "texto": equipo.text
                        }
                    else:
                        print("No se pudo encontrar el elemento de posición")
                except Exception as e:
                    print(f"Error al extraer posición: {str(e)}")
        
        print(f"No se encontró el equipo '{nombre_equipo}'")
        return None
    
    except Exception as e:
        print(f"Error general: {str(e)}")
        return None
    
    finally:
        if driver:
            driver.quit()
```

```python
# Probando la función mejorada
info = obtener_posicion_equipo_urba("Universitario de La Plata")

if info:
    print(f"\n🏉 RESULTADO FINAL 🏉")
    print(f"📍 {info['equipo']} está en la posición {info['posicion']}")
    print(f"📝 Información completa: {info['texto']}")
else:
    print("\n❌ No se pudo encontrar información del equipo.")
    
# También podemos probar con un equipo alternativo para verificar
print("\n--- Probando con otro equipo ---")
otro_info = obtener_posicion_equipo_urba("San Luis")

if otro_info:
    print(f"📍 {otro_info['equipo']} está en la posición {otro_info['posicion']}")
    print(f"📝 Información completa: {otro_info['texto']}")
else:
    print("❌ No se encontró el equipo alternativo.")
```
